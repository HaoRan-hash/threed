```python
import matplotlib.pyplot as plt
```

```python
x = [8, 16, 32, 64, 128, 256]
mious = [65.1, 65.5, 65.9, 66.1, 66.7, 66.8]
oas = [89.3, 89.5, 89.7, 89.7, 89.7, 89.7]
maccs = [71.5, 71.8, 72.3, 72.9, 73.3, 73.5]
```

```python
plt.plot(x, mious, marker='o', color='r')
plt.tick_params(labelsize=10)
plt.savefig('./miou.png', dpi=800)
```

```python
plt.plot(x, oas, marker='o', color='g')
plt.savefig('./oa.png', dpi=800)
```

```python
plt.plot(x, maccs, marker='o', color='b')
plt.savefig('./macc.png', dpi=800)
```
